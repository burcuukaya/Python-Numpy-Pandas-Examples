```python
#################Lambda ve Apply^
#Apply:satır ya da sutunlarda otomatik olarak fonksiyonu calıştırır
df["age2"]=df["age"]*2
df["age3"]=df["age"]*5

#içinde "age" geçen değişkenleri 10 a böl:
df[["age", "age2", "age3"]].apply(lambda x: x/10).head(3)
```

```python
#loc ile satırlardan tümü, sutunlardan yaş ı barındıranı seç : contains
import pandas as pd
import seaborn as sns
df=sns.load_dataset("titanic")
df["age2"]=df["age"]*2
df["age3"]=df["age"]*5
df.loc[:, df.columns.str.contains("age")].apply(lambda x: x / 10).head(3)
```

```python
#İçinde age olan değişkenleri seç,tüm gözlem birimlerinden ortalamayı çıkarıp std a böl:
df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x - x.mean()) / x.std()).head(3)
```

```python
#Fonksiyonu dışarda tanımşayıp uygulamak :
def standard_scaler(col_name):
    return (col_name - col_name.mean())/ col_name.std()
df.loc[:, df.columns.str.contains("age")].apply(standard_scaler).head(3)
```

```python
#"age" içeren değişkenlere yaptığımız bu değişiklikleri kaydetmek istiyoruz
df.loc[:, ["age", "age2", "age3"]]=df.loc[:, df.columns.str.contains("age")].apply(standard_scaler)
```

```python
#age içereni otomatik seçmek :

df.loc[:, df.columns.str.contains("age")]=df.loc[:, df.columns.str.contains("age")].apply(standard_scaler)
df.head()
```

```python
# Görev 23: total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.
import pandas as pd
import seaborn as sns
df=sns.load_dataset("tips")
df["total_bill_tip_sum"]= df["total_bill"] + df["tip"]
new_df = df.sort_values("total_bill_tip_sum", ascending=False)[:30].reset_index(drop=True)
df.head()
```
